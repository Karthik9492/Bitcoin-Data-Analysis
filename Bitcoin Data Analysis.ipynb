# Log Model
Set 'Open time' as the index to make it easier for time series analysis and plotting

```python
bit_coin1d_log = bit_coin1d.set_index('Open time')
```

```python
bit_coin1d_log['Close'].plot()
```

Left Plot: Raw closing prices over time.

Right Plot: Log-transformed closing prices to better visualize growth and reduce skew from large spikes.

```python
plt.figure(figsize= (20,6))

plt.subplot(1,2,1)
bit_coin1d_log['Close'].plot()
plt.title("No Scaling")

plt.subplot(1,2,2)
np.log1p(bit_coin1d_log['Close']).plot()
plt.title("Log Scaling")
```

The left plot shows raw Open price values.

The right plot applies a log transformation (log1p) which compresses large values and reveals relative changes better.

```python
plt.figure(figsize= (20,6))

plt.subplot(1,2,1)
bit_coin1d_log['Open'].plot()
plt.title("No Scaling")

plt.subplot(1,2,2)
np.log1p(bit_coin1d_log['Open']).plot()
plt.title("Log Scaling")
```

# Resample method
Performunig Analysis on Opening price on Yearly, Quarterly, Monthly

```python
bit_coin1d_log['Open'].resample('QE').mean().plot()
```

```python
bit_coin1d_log['Open'].resample('ME').mean().plot()
```

# Daily change in Closing & Opening price of stock
pct_change() calculates the percentage change between the current and previous value. Multiplying by 100 converts it into a percentage.

iplot() creates an interactive line chart with the daily price changes.

go_offline() ensures it runs without needing an internet connection.

```python
bit_coin1d_log['Close_pct_change'] = bit_coin1d_log['Close'].pct_change()*100
bit_coin1d_log['Open_pct_change'] = bit_coin1d_log['Open'].pct_change()*100
```
